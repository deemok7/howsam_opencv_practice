### <span style="font-size: 0.9em;">üüß</span> **Image channels**

```python
img3_bgr = cv2.imread("imgs/basketball.png")
if img3_bgr is None:
    print("Error")

print(img3_bgr.shape)

plt.Figure(figsize=(3, 3))
plt.imshow(img3_bgr)
plt.axis("off")


cv2.imshow("", img3_bgr)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
b = img3_bgr[..., 0]
g = img3_bgr[..., 1]
r = img3_bgr[..., 2]
print(r.shape)

img3_rgb = np.stack((r, g, b), axis=2)
print(img3_bgr.shape)

plt.Figure(figsize=(3, 3))
plt.imshow(img3_rgb)
plt.axis("off")

cv2.imshow("", img3_rgb)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img3_rgb = cv2.cvtColor(img3_bgr, cv2.COLOR_BGR2RGB)
print(img3_bgr.shape)

plt.Figure(figsize=(3, 3))
plt.imshow(img3_rgb)
plt.axis("off")


cv2.imshow("", img3_rgb)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img3_rgb = cv2.cvtColor(img3_bgr, cv2.COLOR_BGR2RGB)
print(img3_bgr.shape)

plt.Figure(figsize=(3, 3))
plt.imshow(img3_rgb)
plt.axis("off")

tmp_bgr = cv2.cvtColor(img3_rgb, cv2.COLOR_RGB2BGR)
cv2.imshow("", tmp_bgr)
cv2.waitKey()
cv2.destroyAllWindows()
del tmp_bgr
```

```python
img3_gray = cv2.cvtColor(img3_bgr, cv2.COLOR_BGR2GRAY)
print(img3_gray.shape)

plt.Figure(figsize=(3, 3))
plt.imshow(img3_gray)
# plt.imshow(np.stack((img3_gray, img3_gray, img3_gray), axis=2))
plt.axis("off")

cv2.imshow("", img3_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

### <span style="font-size: 0.9em;">üüß</span> **Numpy &#x21D4; OpenCV**

```python
img = np.ones((255, 255), dtype=np.uint8)
print(img.shape, np.min(img), np.max(img))

cv2.imshow("", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = 127 * np.ones((255, 255), dtype=np.uint8)
print(img.shape, np.min(img), np.max(img))

cv2.imshow("", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = 255 * np.ones((255, 255), dtype=np.uint8)
print(img.shape, np.min(img), np.max(img))

cv2.imshow("", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = 255 * np.ones((255, 255), dtype=np.uint8)
print(img.shape, np.min(img), np.max(img))

img[:, ::5] = 0

cv2.imshow("", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = 255 * np.ones((255, 255), dtype=np.uint8)
print(img.shape, np.min(img), np.max(img))

img[:, ::10] = 0
img[::10, :] = 0

cv2.imshow("", img)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imwrite("results/grid.bmp",img)
```
