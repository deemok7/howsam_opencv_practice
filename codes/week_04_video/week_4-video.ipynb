### <span style="font-size: 0.9em;">üüß</span> **More on `cv2.VideoCapture()`**

#### <span style="font-size: 0.9em;">üü®</span> **Get video from url**

```python
cap = cv2.VideoCapture("https://dl.howsam.org/opencv/data/table-tennis.mp4")

print(f"Height: {cap.get(cv2.CAP_PROP_FRAME_HEIGHT)}, Width: {cap.get(cv2.CAP_PROP_FRAME_WIDTH)}")
print(f"Frame Count: {cap.get(cv2.CAP_PROP_FRAME_COUNT)}")
print(f"FPS: {cap.get(cv2.CAP_PROP_FPS)}")


while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Frame", frame)

    key = cv2.waitKey(40)
    if key == ord("q"):
        break
    
cv2.destroyAllWindows()
cap.release()
```
