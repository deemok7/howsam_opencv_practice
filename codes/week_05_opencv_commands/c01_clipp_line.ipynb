```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((500, 500, 3), dtype=np.uint8)

rect = (50, 50, 300, 300)

line_p1 = (70, 100)
line_p2 = (400, 600)


cv2.rectangle(
    image, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (255, 0, 0), 1
)
cv2.line(image, line_p1, line_p2, (0, 0, 255), 1)

clipped, clipped_line_p1, clipped_line_p2 = cv2.clipLine(rect, line_p1, line_p2)


if clipped:
    print("✅ Line after clipping:", clipped_line_p1, clipped_line_p2)
    cv2.line(image, clipped_line_p1, clipped_line_p2, (0, 255, 0), 2)
else:
    print("❌ Line is completely outside the rectangle.")


rect = (50, 50, 300, 300)

line_p1 = (10, 200)
line_p2 = (500, 600)


cv2.rectangle(
    image, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (255, 0, 0), 1
)
cv2.line(image, line_p1, line_p2, (0, 0, 255), 1)

clipped, clipped_line_p1, clipped_line_p2 = cv2.clipLine(rect, line_p1, line_p2)

if clipped:
    print("✅ Line after clipping:", clipped_line_p1, clipped_line_p2)
    cv2.line(image, clipped_line_p1, clipped_line_p2, (0, 255, 0), 2)
else:
    print("❌ Line is completely outside the rectangle.")


image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.title("Line Clipping Visualization")
plt.axis("on")
plt.show()
```
